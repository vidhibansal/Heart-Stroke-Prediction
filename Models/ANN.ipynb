```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
# fix random seed for reproducibility
numpy.random.seed(7)
seed = 7
```

```python
# load dataset
dataset = numpy.loadtxt("dataset.csv", delimiter=",",skiprows=1)
# split into input (X) and output (Y) variables
X = dataset[:,1:15]
Y = dataset[:,15]
# split into 67% for train and 33% for test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=seed)
```

```python
#We are using fully-connected network structure with three layers. Fully connected layers are defined using the Dense class.
#We will use the rectifier (‘relu‘) activation function on the first two layers and the sigmoid function in the output layer.


# create model
model = Sequential()
model.add(Dense(12, input_dim=14, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# we will run for a small number of iterations and use a relatively small batch size of 10. 

# Fit the model
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=100, batch_size=10)
```

```python
# evaluate the model 
scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
expected = Y
# Make predictions
predictions = model.predict(X)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

```python
#Print the confusion matrix
results = confusion_matrix(expected, rounded)
print(results)
```
